# 08 - Basic RAG

This notebook shows the base question-answering flow: retrieve chunks first, inspect them, build context, and only then call the model.

```python
from llmkit.config.settings import get_settings
from llmkit.rag.answer import answer_with_context
from llmkit.rag.vectorstores import retrieve_chunks

settings = get_settings()
VECTORSTORE_PATH = settings.vectorstore_dir / "local_docs"
QUESTION = "Who founded Insurellm?"
```

## 1. Retrieve chunks manually

```python
retrieved_chunks = retrieve_chunks(QUESTION, persist_directory=VECTORSTORE_PATH, k=4)
for index, chunk in enumerate(retrieved_chunks, start=1):
    print(f"\n--- Chunk {index} ---")
    print(chunk.metadata)
    print(chunk.page_content[:500])
```

## 2. Use the reusable answering helper

```python
result = answer_with_context(QUESTION, retrieved_chunks)
print("SYSTEM PROMPT:\n", result.system_prompt)
print("\nUSER PROMPT:\n", result.user_prompt)
print("\nANSWER:\n", result.answer)
print("\nSOURCES:\n", result.sources)
```
